```python
import wave
import io

# List all objects in the specified prefix
prefix = '2025-04-11T08-08-20.981Z/'
response = s3.list_objects_v2(Bucket='dmh-prod1-deploy-bucket', Prefix=prefix)

if 'Contents' in response:
    wav_files = [obj['Key'] for obj in response['Contents'] if obj['Key'].endswith('.wav')]
    print(f"Found {len(wav_files)} .wav files in {prefix}.")
else:
    print(f"No files found in {prefix}.")
    wav_files = []

# Combine all .wav files
output_wav = io.BytesIO()
with wave.open(output_wav, 'wb') as output_wave:
    for idx, wav_file in enumerate(wav_files):
        response = s3.get_object(Bucket='dmh-prod1-deploy-bucket', Key=wav_file)
        wav_data = io.BytesIO(response['Body'].read())
        
        with wave.open(wav_data, 'rb') as input_wave:
            if idx == 0:
                # Set parameters for the output file based on the first file
                output_wave.setparams(input_wave.getparams())
            else:
                # Ensure all files have the same parameters
                assert input_wave.getnchannels() == output_wave.getnchannels()
                assert input_wave.getsampwidth() == output_wave.getsampwidth()
                assert input_wave.getframerate() == output_wave.getframerate()
            
            # Write frames to the output file
            output_wave.writeframes(input_wave.readframes(input_wave.getnframes()))

output_wav.seek(0)
print("All .wav files have been combined into one.")
```

```python
# Save the combined WAV file to disk
output_file_path = 'combined_audio.wav'
with open(output_file_path, 'wb') as f:
    f.write(output_wav.getvalue())

print(f"Combined audio saved to {output_file_path}")
```
